```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
import nltk
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from sklearn.cluster import MiniBatchKMeans

# Load SBERT model
model = SentenceTransformer('paraphrase-MiniLM-L3-v2')

# Function to read and preprocess text data
def read_and_preprocess(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    sentences = nltk.sent_tokenize(text)  # Tokenize into sentences
    return [sentence.strip() for sentence in sentences if sentence.strip()]

# Load datasets
history_sentences = read_and_preprocess('src/data/wikipedia_timeline/1940s.txt')
movies_sentences = read_and_preprocess('src/data/summaries_per_decade/1940.txt')

history_sentences = history_sentences[:500]  # Use the first 500 sentences as a test
movies_sentences = movies_sentences[:100]   # Use the first 200 sentences as a test

# Encode sentences into embeddings
history_embeddings = model.encode(history_sentences, batch_size=32, convert_to_tensor=True)
movies_embeddings = model.encode(movies_sentences, batch_size=32, convert_to_tensor=True)


# Compute pairwise similarity matrix
similarity_matrix = util.pytorch_cos_sim(history_embeddings, movies_embeddings).cpu().numpy()

# Perform clustering on history dataset
num_clusters = min(20, len(history_embeddings))  # Example: max 20 clusters
cluster_model = MiniBatchKMeans(n_clusters=num_clusters)
history_clusters = cluster_model.fit_predict(history_embeddings.cpu().numpy())

# Cluster sentences from history into semantic fields
clustered_history = {i: [] for i in range(max(history_clusters) + 1)}
for idx, cluster_id in enumerate(history_clusters):
    clustered_history[cluster_id].append(history_sentences[idx])

# Compute average similarity per cluster
topic_similarities = []
cluster_avg_similarities = []
for cluster_id, sentences in clustered_history.items():
    cluster_emb = model.encode(sentences, convert_to_tensor=True).mean(dim=0)  # Average embedding for the cluster
    similarities = util.pytorch_cos_sim(cluster_emb, movies_embeddings).cpu().numpy()
    cluster_avg_similarities.append(similarities.mean(axis=1))  # Store average similarities
    topic_similarities.append(similarities.max())  # Store max similarity for the bar plot

# Save bar plot data (Maximum similarity per cluster)
bar_plot_data = pd.DataFrame({
    'Cluster ID': list(clustered_history.keys()),
    'Max Similarity': topic_similarities
})
bar_plot_data.to_csv('bar_plot_data.csv', index=False)

# Save heatmap data (Average similarity matrix)
heatmap_data = pd.DataFrame(
    cluster_avg_similarities,
    index=[f"Cluster {i}" for i in range(len(clustered_history))],
    columns=[f"Movie {i+1}" for i in range(len(movies_sentences))]
)
heatmap_data.to_csv('heatmap_data.csv')

# Visualization: Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=bar_plot_data['Cluster ID'], y=bar_plot_data['Max Similarity'], palette="viridis")
plt.xlabel('Cluster ID (History)')
plt.ylabel('Max Similarity with Movie Summaries')
plt.title('Maximum Correlation Between History Clusters and Movie Summaries')
plt.show()

# Visualization: Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(cluster_avg_similarities, annot=True, fmt=".2f", cmap="coolwarm", 
            xticklabels=[f"Movie {i+1}" for i in range(len(movies_sentences))],
            yticklabels=[f"Cluster {i}" for i in range(len(clustered_history))])
plt.title('Semantic Correlation Heatmap: History Clusters vs Movie Summaries')
plt.xlabel('Movie Summaries')
plt.ylabel('History Clusters')
plt.show()
```

```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
import numpy as np
import nltk
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Load SBERT model
model = SentenceTransformer('paraphrase-MiniLM-L3-v2')

# Function to read and preprocess text data
def read_and_preprocess(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    sentences = nltk.sent_tokenize(text)  # Tokenize into sentences
    return [sentence.strip() for sentence in sentences if sentence.strip()]

# Load summaries per decade
summaries = pd.read_csv('src/data/summaries_per_decade.txt', sep='\t', header=None)
summary_data = {f"summaries_{1900 + i * 10}": summaries[0][i] for i in range(11)}

# List historical data files (assumes naming convention like 'history_1900.txt', 'history_1910.txt', ...)
historical_files = [f"src/data/history_{1900 + i * 10}.txt" for i in range(11)]

# Initialize data storage for results
bar_plot_results = []


heatmap_results = {}

# Process each decade
for i, decade in enumerate(range(1900, 2011, 10)):
    # Load history and summary data for the decade
    history_file = historical_files[i]
    history_sentences = read_and_preprocess(history_file)
    
    summary_text = summary_data[f"summaries_{decade}"]
    summaries_sentences = nltk.sent_tokenize(summary_text)
    
    # Get embeddings
    history_embeddings = model.encode(history_sentences, convert_to_tensor=True)
    summaries_embeddings = model.encode(summaries_sentences, convert_to_tensor=True)
    
    # Compute pairwise similarity matrix
    similarity_matrix = util.pytorch_cos_sim(history_embeddings, summaries_embeddings).cpu().numpy()
    
    # Perform clustering on history dataset
    cluster_model = AgglomerativeClustering(n_clusters=None, distance_threshold=1.5, affinity='cosine', linkage='average')
    history_clusters = cluster_model.fit_predict(history_embeddings.cpu().numpy())
    
    # Cluster sentences from history into semantic fields
    clustered_history = {j: [] for j in range(max(history_clusters) + 1)}
    for idx, cluster_id in enumerate(history_clusters):
        clustered_history[cluster_id].append(history_sentences[idx])
    
    # Compute average similarity per cluster
    topic_similarities = []
    cluster_avg_similarities = []
    for cluster_id, sentences in clustered_history.items():
        cluster_emb = model.encode(sentences, convert_to_tensor=True).mean(dim=0)  # Average embedding for the cluster
        similarities = util.pytorch_cos_sim(cluster_emb, summaries_embeddings).cpu().numpy()
        cluster_avg_similarities.append(similarities.mean(axis=1))  # Store average similarities
        topic_similarities.append(similarities.max())  # Store max similarity for the bar plot
    
    # Store bar plot data
    bar_plot_results.append({
        'Decade': decade,
        'Cluster IDs': list(clustered_history.keys()),
        'Max Similarities': topic_similarities
    })
    
    # Store heatmap data
    heatmap_results[decade] = np.array(cluster_avg_similarities)
    
    # Save CSV files for each decade
    pd.DataFrame({
        'Cluster ID': list(clustered_history.keys()),
        'Max Similarity': topic_similarities
    }).to_csv(f'results/bar_plot_data_{decade}.csv', index=False)
    
    heatmap_df = pd.DataFrame(
        cluster_avg_similarities,
        index=[f"Cluster {j}" for j in range(len(clustered_history))],
        columns=[f"Summary {k+1}" for k in range(len(summaries_sentences))]
    )
    heatmap_df.to_csv(f'results/heatmap_data_{decade}.csv')

    # Visualization: Bar Plot
    plt.figure(figsize=(10, 6))
    sns.barplot(x=list(clustered_history.keys()), y=topic_similarities, palette="viridis")
    plt.xlabel('Cluster ID (History)')
    plt.ylabel('Max Similarity with Movie Summaries')
    plt.title(f'Maximum Correlation: History vs Movies ({decade})')
    plt.show()

    # Visualization: Heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(heatmap_results[decade], annot=True, fmt=".2f", cmap="coolwarm",
                xticklabels=[f"Summary {k+1}" for k in range(len(summaries_sentences))],
                yticklabels=[f"Cluster {j}" for j in range(len(clustered_history))])
    plt.title(f'Semantic Correlation Heatmap: History vs Movies ({decade})')
    plt.xlabel('Movie Summaries')
    plt.ylabel('History Clusters')
    plt.show()
```
