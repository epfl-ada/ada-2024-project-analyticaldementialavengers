```python
import requests
from bs4 import BeautifulSoup
import csv

# Step 1: Fetch the page content and get only the text
url = "https://en.wikipedia.org/wiki/Timeline_of_the_20th_century"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
text_content = soup.get_text()

# Step 2: Initialize a list to store events and set up variables
events = []
current_year = None

# Step 3: Process the content line by line
for line in text_content.splitlines():
    line = line.strip()  # Remove extra whitespace

    # Check if the line is a year (four-digit number)
    if line.isdigit() and len(line) == 4:
        current_year = line  # Update the current year
        print(f"Found year: {current_year}")  # Debug: Print the year found

    # Check if the line starts with a dash (indicating an event entry)
    elif current_year and line.startswith("•"):
        # Extract event description by splitting at ": "
        parts = line.split(": ", 1)
        if len(parts) > 1:
            event_description = parts[1].strip()  # Take only the event description
            events.append((current_year, event_description))
            print(f"Recorded event: Year={current_year}, Event={event_description}")  # Debug

# Step 4: Save events to a CSV file
with open("prova_timeline.csv", "w", newline='', encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Year", "Event"])  # Write header
    writer.writerows(events)  # Write all events

print("CSV file created successfully with timeline events.")
```
