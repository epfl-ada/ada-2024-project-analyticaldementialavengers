```python
import pandas as pd 
import numpy as np
```

```python
# source : https://ourworldindata.org/grapher/median-age?tab=table&time=earliest..2012&country=JPN~IND~GBR~USA~FRA~ITA~CAN~DEU~ARG~HKG~AUS~ESP
median_age=pd.read_csv('median-age.csv')
median_age.head()
```

```python
# renaming the columns
median_age.columns=['Country','Code','Year','Median age', 'Median age prediction']

# removing rows corresponding to years after 2013
index=median_age[median_age['Year']>2013].index
median_age=median_age.drop(index)

# removing the 'Code' and 'Median age prediction' columns
median_age=median_age.drop('Code',axis=1)
median_age=median_age.drop('Median age prediction',axis=1)
median_age
```

```python
dict_us={'United States': median_age[median_age['Country']=='United States']['Median age']}
df_us=pd.DataFrame(dict_us)
df_us
```

```python
# creating a df of the median age between 1950 and 2013, for the 11 countries with more than 1000 films in the database
country_list=['United States', 'India', 'United Kingdom', 'France', 'Italy', 'Japan', 'Canada', 'Germany', 'Argentina', 'Hong Kong', 'Spain', 'Australia']
dict_countries={}
for country in ['United States', 'India', 'United Kingdom', 'France', 'Italy', 'Japan', 'Canada', 'Germany', 'Argentina', 'Hong Kong', 'Spain', 'Australia']:
    dict_countries[country]=median_age[median_age['Country']==country]['Median age'].tolist() # extracting the median age data for the given country
countries_df=pd.DataFrame(dict_countries)
years_df=pd.DataFrame(median_age[median_age['Country']=='Afghanistan']['Year']) # extracting the column year
median_age_df=pd.concat([years_df,countries_df],axis=1) # creating a df with a column for the year and a column for each country
median_age_df
```
