```python
import pandas as pd 
import numpy as np
import re
names=['wiki_ID','freebase_ID','name','release_date','revenue','runtime','languages','countries','genres']
movie_metadata=pd.read_csv('../Data/movie.metadata.tsv', sep='\t',names=names)
movie_metadata
```

```python
# extracting a list of all countries of a movie from a string in the column 'countries' in movie_metadata
def extract_countries(string):
    between_quotes = re.findall(r'"(.*?)"', string) # extracting all words from the string that are between quotes
    for word in between_quotes:
        if word[0]=='/':
            between_quotes.remove(word) # removing all words between quotes that start with /, as they are not country names
    return between_quotes # we are left with only words corresponding to country names
```

```python
# counting the number of countries of a movie from a string in the column 'countries' in movie_metadata
def count_countries(string):
    country_number=string.count(',') +1 # counting the number of comas in the string, as they are separators between different countries
    return country_number # the number of comas + 1 corresponds to the number of countries
```

```python
movie_metadata['country_number']=movie_metadata['countries'].astype(str).apply(count_countries) # new column in movie_metadata with the number of countries for each film
movie_metadata['country_list']=movie_metadata['countries'].astype(str).apply(extract_countries) # new column in movie_metadata with a list of strings of all country names for each film
```

```python
# creating a list of all country names of all films combined, to count the number of occurrences of each country
total_country_list=[]
for list in movie_metadata['country_list']: # iterating over the lists of countries for all films
    for element in list: # iterating over the countries in a given list of countries
        total_country_list.append(element) # adding each country name to the list of all countries
total_country_list=pd.DataFrame({'country':total_country_list})
country_distribution_df=total_country_list.value_counts().reset_index() # df of all countries and their number of occurrence in all films
country_distribution_df[country_distribution_df['count']>1000] # displaying only countries that are present in more than 1000 films
```
